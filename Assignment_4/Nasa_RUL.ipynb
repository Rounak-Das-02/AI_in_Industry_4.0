```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,r2_score
```

```python
columns=["id","cycle","op1","op2","op3","sensor1","sensor2","sensor3","sensor4","sensor5","sensor6","sensor7","sensor8",
         "sensor9","sensor10","sensor11","sensor12","sensor13","sensor14","sensor15","sensor16","sensor17","sensor18","sensor19"
         ,"sensor20","sensor21","sensor22","sensor23"]
```

```python
#reading the dataset
train=pd.read_csv("train_FD001.txt",sep=" ",names=columns)
test=pd.read_csv("test_FD001.txt",sep=" ",names=columns)
test_result=pd.read_csv("RUL_FD001.txt",sep=" ",header=None)
```

### Data Preprocessing

```python
test_result.head()
```

```python
train.info() # to get info of your dataset ex: null values,datatype..,
```

```python
test_result.columns=["rul","null"]
test_result.head()
```

```python
test_result.drop(["null"],axis=1,inplace=True)
test_result["id"]=test_result.index+1
test_result.head()
```

```python
rul=pd.DataFrame(test.groupby('id')['cycle'].max()).reset_index()
rul.columns=['id','max']
```

```python
rul.head()
```

```python
train.head()
```

```python
test_result['rul_failed']=test_result["rul"]+rul["max"]
test_result.head()
```

```python
test_result.drop(['rul'],axis=1,inplace=True)
```

```python
test=test.merge(test_result,on=['id'],how='left')
```

```python
test["remaining_cycle"]=test["rul_failed"]-test["cycle"]
test.head()
```

```python
df_train=train.drop(["sensor22","sensor23"],axis=1)
df_test=test.drop(["sensor22","sensor23"],axis=1)
```

```python
df_train["remaining_cycle"]= df_train.groupby(['id'])['cycle'].transform(max)-df_train['cycle']
```

```python
x=df_train.drop(["id","op3","sensor1","sensor5","sensor6","sensor10","sensor16","sensor18","sensor19","remaining_cycle"],axis=1)
y=df_train['remaining_cycle']
print('x shape : ',x.shape)
print('y shape : ',y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=3)
print('X_train shape : ',x_train.shape)
print('X_test shape : ',x_test.shape)
print('y_train shape : ',y_train.shape)
print('y_test shape : ',y_test.shape)
```

### Applying Various Regression Approachs

```python
import xgboost as xgb
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# Linear Regression
LR_REGRESSOR = LinearRegression()
LR_REGRESSOR.fit(x_train, y_train)
y_pred_lr = LR_REGRESSOR.predict(x_test)

LR_MAE = mean_absolute_error(y_test, y_pred_lr)
LR_MSE = mean_squared_error(y_test, y_pred_lr)
LR_RMSE = np.sqrt(LR_MSE)
r2_lr = r2_score(y_test, y_pred_lr)

print(f"Model: Linear Regression")
print("MAE : " , LR_MAE)
print("MSE : " , LR_MSE)
print("RMSE : ", LR_RMSE)
print("R2 : ", r2_lr)
```

```python
#RandomForestRegressor
rf_regressor=RandomForestRegressor(n_jobs=-1,n_estimators=600)
rf_regressor.fit(x_train,y_train)
y_pred_rf=rf_regressor.predict(x_test)


MAE=mean_absolute_error(y_test,y_pred_rf)
MSE=mean_squared_error(y_test,y_pred_rf)
RMSE=np.sqrt(MSE)
r2=r2_score(y_test,y_pred_rf)
print(f"Model: RandomForestRegressor")
print("MAE : " , MAE)
print("MSE : " , MSE)
print("RMSE : ", RMSE)
print("R2 : ", r2)
```

```python
#GradientBoostingRegressor
gb_regressor = GradientBoostingRegressor(n_estimators=600)
gb_regressor.fit(x_train, y_train)
y_pred_gb = gb_regressor.predict(x_test)


MAE = mean_absolute_error(y_test, y_pred_gb)
MSE = mean_squared_error(y_test, y_pred_gb)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test, y_pred_gb)
print(f"Model: GradientBoost Regressor")
print("MAE : " , MAE)
print("MSE : " , MSE)
print("RMSE : ", RMSE)
print("R2 : ", r2)
```

```python
#XGBoost
gb_regressor = xgb.XGBRegressor()
gb_regressor.fit(x_train, y_train)
y_pred_gb = gb_regressor.predict(x_test)


MAE = mean_absolute_error(y_test, y_pred_gb)
MSE = mean_squared_error(y_test, y_pred_gb)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test, y_pred_gb)
print(f"Model: XGBoost Regressor")
print("MAE : " , MAE)
print("MSE : " , MSE)
print("RMSE : ", RMSE)
print("R2 : ", r2)
```
